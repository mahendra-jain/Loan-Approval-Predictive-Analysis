```python
loan_data_copy['loan_percent_income (%)'] = loan_data_copy['loan_percent_income (%)']* 100
```

```python
loan_data_ml = loan_data_copy.copy()
```

```python
# Machine Learning Code
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Encode categorical columns
categorical_cols = ['person_gender', 'person_education', 'person_home_ownership',
                    'loan_intent', 'previous_loan_defaults_on_file', 'loan_status']

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    loan_data_ml[col] = le.fit_transform(loan_data_ml[col])
    label_encoders[col] = le
```

```python
label_encoders
```

```python
x =loan_data_ml.drop('loan_status',axis=1)
y= loan_data_ml['loan_status']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
